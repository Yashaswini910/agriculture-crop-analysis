```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation Heatmap
df = pd.read_csv('agri_project.csv')
corr = df[['Yield_t_per_ha', 'Monsoon_rainfall',
           'Rainfall_mm', 'MSP_Price']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# District performance vs State average
district_avg = df.groupby('District')['Yield_t_per_ha'].mean()
state_avg = df['Yield_t_per_ha'].mean()
district_avg.plot(kind='bar', color='skyblue')
plt.axhline(state_avg, color='red', linestyle='--', label='State Average')
plt.title('District Yield vs State Average')
plt.legend()
plt.show()
```
